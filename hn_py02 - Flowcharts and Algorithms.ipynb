## Flowharts
One way of representing algorithms is to use flow charts, also called flow diagrams. They are a useful way of planning how a computer program
might work, and show others your thinking. A flow chart shows the key points in an algorithm:

1.  The start and end

2.  The order in which the sequences of instructions are performed

3.  The points where inputs and outputs occur

4.  The points where decisions are made about what to do next

5.  A sequence of many instructions that does not involve any of these, key points may be represented as a single rectangular box.

Flowcharts use a variety of standard flowchart symbols to represent different elements, and arrows to show the flow or direction. These
shapes are formally agreed standards (BS 4058:1987, ISO 5807:1985). The most commonly used symbols are:

## Pseudocode

Another way of expressing an algorithm without conforming
to specific syntactic rules. By learning to read and write pseudocode,
learners can more easily communicate ideas and concepts to other
programmers, even though they may be using completely different
languages. What’s more, algorithmic solutions to many problems are often
provided in pseudocode on sites such as StackOverflow and Wikipedia,
meaning an ability to translate between pseudocode and a given
programming language is a valuable skill.

There is no such thing as correct pseudocode, although there are a few
generally accepted notations that are widely understood by programmers
(e.g. x = 10 to create a variable called x with a value of 10), no
matter which language they use.

FizzBuzz is a common programming challenge with hundreds of different
solutions in many languages. The challenge is as follows. For every
number from 1 to 100, output Fizz if the number is divisible by 3,
output Buzz if the number is divisible by 5, and output FizzBuzz if the
number is divisible by both 3 and 5. If none of these conditions match,
then just output the number. Let’s have a look at one possible
pseudocode interpretation of this challenge:

`FOR i \<-- 1 TO 100 DO`

`IF i is divisible by 3 AND i is divisible by 5 THEN`

`OUTPUT "FizzBuzz"`

`ELSE IF i is divisible by 3 THEN`

`OUTPUT "Fizz"`

`ELSE IF i is divisible by 5 THEN`

`OUTPUT "Buzz"`

`ELSE`

`OUTPUT i`

### Stepwise Refinement

One of the oldest methods still used today in computer programming is
Stepwise refinement. Stepwise refinement refers to the progressive
refinement in small steps of a program specification into a program.
Sometimes, it is called top-down design. The term stepwise refinement
was used first in the paper titled Program Development by Stepwise
Refinement by Niklaus Wirth, the author of the programming language
Pascal and other major contributions to software design and software
engineering, in the Communications of the ACM, Vol. 14 (4), 1971, pp.
221-227.

Wirth said, "It is here considered as a sequence of design decisions
concerning the decomposition of tasks into subtasks and of data into
data structures." We use the term to describe the development of a
method from its specification. We will see small steps, like breaking a
high-level statement into a sequence of statements. Consider the
following two examples:

#### 1\. Changing time formats

Translating a time of day from 24-hour format into conventional AM-PM
format. Two stages of stepwise refinement are used: firstly you need to
decide if the time is AM or PM, so is the number greater than 12? This
breaks the problem into two parts, the AM part should be easy now you
just add the AM to the end of the time and drop the zero, so 09:00
becomes 9am, 0815 becomes 8:15am and so on. The next step is to convert
the 24-hour times to 12-hour times, this can be achieved by taking 12
away from the first number so 13-12 = 1, that's 1pm, 18 -12 is 6, that's
6pm.

#### 2\. Anglicising an number

Anglicise means to adapt (a foreign word, name, or phrase) to English
usage. According to the dictionary, the first known use of the word was
in 1710. Stepwise refinement could be used to develop a method to
anglicise numbers ---to turn integers like 1710 into words: one thousand
seven hundred and ten. It's amazing how stepwise refinement results in a
simple method, and the development incorporates recursion in a natural
way. We would have to take each column in the system and convert it to
words, then put them together, so 1 thousand, 7 hundreds, and 1 tens
would convert to one thousand seven hundred and ten.

A top-down approach (also known as stepwise design) is essentially the
breaking down of a system to gain insight into the subsystems that make
it up. In a top-down approach an overview of the system is formulated,
specifying but not detailing any first-level subsystems. Each subsystem
is then refined in yet greater detail, sometimes in many additional
subsystem levels, until the entire specification is reduced to base
elements. Once these base elements are recognised then we can build
these as computer modules. Once they are built we can put them together,
making the entire system from these individual components.

To show you an example of Top-Down design we'll take the act of making
pancakes. Starting from the top we have the task to ***Make pancakes.***
Saying that by itself isn't enough to actually make any pancakes, we
need to break the task down:

#### Make some pancakes

* Organise Kitchen

* Make Pancakes

* Serve

#### Each of these tasks can then be broken down further:

* Organise Kitchen

* Clean surfaces

* Get out mixing bowl, whisk, spoon, sieve

* Get out plain flour, salt, eggs, milk, butter

* Put on apron

* Make Pancakes

* Sift salt and flour into bowl

* Break eggs into bowl

* Whisk

* Add water and milk

* Add butter

* Whisk

* Cook

* Serve

#### And each of these tasks can be broken down further, let us take a look at steps in the cooking task:

* Get pan to temperature

* Pour batter in

* Spread batter to edges

* Use plastic spatula to check bottom of pancake

* When brown, flip

* Use plastic spatula to check bottom of pancake

* When brown finish

We can break down some of these tasks even further. So starting at a single point, the creation of a pancake, we have broken down the whole process into a series of individual tasks.

### Algorithmic Efficiency

One of the simplest methods for guessing a number is called ***exhaustive enumeration.*** Let's say we want to find a value and the value is 10, using exhaustive enumeration we guess 1 and then check if this is 10, if not we add 1 and then check again. Eventually we will get to the right answer. This works just fine for a small number but let's say we had to guess 100 or 1,000,000 - this would take a much longer time.

Can you think of a more efficient way to find a number? In topic 1 we learned about the guess and check method - of course you could take a
guess. Let's say we guess the right number first time - then of course that is more efficient than exhaustive enumeration - but what if the
guess was always wrong? It could take forever to find the right number.

Another method of guessing a number is called the ***bi-section method***, sometimes called bisection-search or interval halving. This
is a simple form of the bisection method. The rules are to guess a value of 10 are:

Take the endpoint-values (highest and lowest) 1 and 10 and then we divide the highest value by 2 and then round up if the answer is 0.5 or
higher. This should give an answer of five.

Then we check if 5 is the right answer or if it is too high or too low, 5 is too low so we take the guess and add it to the endpoint which gives
us 15, then divide that by two. 5 + 10 = 15 then 15 / 2 is 7.5 - round 7.5 up to 8.

We check if 8 is right and find it's too low so repeat the process 10+8 = 18 divided by 2 gives 9.

Still too low so add 10 to 9 and divide by two gives 9.5 round up and you have 10.

Using this method we got the answer is 4 steps compared to 10 steps in
exhaustive enumeration? At the moment this may seem trivial however if
you had to calculate a very very large number or numbers a lot of times
this method would be much more efficient.

Now let's put this to the test - using a calculator, on your phone or on
a computer try getting to 1000 with the bi-section method and rounding
up - we know exhaustive enumeration would have 1000 steps but how many

2.  Round 250 up to 300 add 300 to 500 and divide by 2

3.  Add 400 to 500 and divide by 2

# Part 2 Syntax

Syntax can be defined as the set of rules or principles that govern the
structure of well formed sentences in a language. A sentence like:

Today I went to visit a friend.

Is a syntactically valid sentence, if i were to send you a message with
this sentence and wrote it like this:

^oday & 12nt t9 vi6it “ friend

You would probably reply and say - what do you mean? That is because
this sentence violates the syntactic rules of the English language,
there is no such word as ^oday & 12nt. In computer programming, the
computer only understands instructions if they are written in a way that
the computer can read. In python to print out a word to the screen we
write it like this:

`\>\>\> print (“Hello World”)`

This will return:

`\>\>\> Hello World`

The computer prints it out because you have given it a syntactically
valid instruction. If you miss out one of the quotation marks or forget
to close a bracket or mis-spell the word print, the computer will return
a syntax error - examples of incorrect print statements include:

`Print (“Hello World”) - print should be in lowercase`

`print (“Hello World” - missing the closing parenthesis`

`print (Hello World”) - missing the opening double quotation marks`

`pirnt (“Hello World”) - the word print is misspelled`

Syntax errors are usually easy for the computer to spot and also are
quite easy to fix, its common to make a spelling mistake or miss out a
bracket and these kind of errors are easily resolved

#### Semantic Errors

A semantic error is different from a syntax error, in such that the
instruction can be syntactically valid and so the computer tries to
interpret this, however the outcome generates something that is
unexpected. Consider the sentence again:

Today I went to visit a friend.

This is a syntactically and semantically valid sentence but let's say we
changed it to:

Today visit went I to friend.

Now this does not break any English syntax laws, so the words are valid
words - however they do not make sense in the order that we have
supplied them. Semantic errors are more difficult for the computer to
spot because often the instructions are valid. Consider the following
sentence:

"John is a married bachelor."

This is grammatically well formed but expresses a meaning that cannot be
true, how can John be a bachelor and also be married?

#### Syntax and Semantics

Just because something is semantically valid in one language - this
doesn't mean it is valid in another. If i asked you to print (“Hello”)
you would say - what do you mean print, do you want me to go to the
printer and print out a piece of paper with the word hello on it? If you
went to a local coffee shop and said “Je voudrais un café” unless the
person spoke French they would say something like "sorry what was that?" What you said
was syntactically and semantically valid in French, but not in English.
